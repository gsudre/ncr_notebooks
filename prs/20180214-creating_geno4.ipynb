Before we merge the boxes we need to remove the family IDs in each box, so that the same NSB is in two different boxes but with different family IDs they don't get included, just one of them:

````bash
n=`cat ../geno3/update_sex.txt | wc -l`;
if [ -e inds_ids.txt ]; then rm inds_ids.txt; fi;
for i in `seq 1 $n`; do echo 1 >> inds_ids.txt; done;
cut -f 2,3 ../geno3/update_sex.txt > tmp.txt;
paste inds_ids.txt tmp.txt > noid_update_sex.txt
tail noid_update_sex.txt
````

````bash
n=`cat ../geno3/update_ids.txt | wc -l`;
if [ -e inds_ids.txt ]; then rm inds_ids.txt; fi;
for i in `seq 1 $n`; do echo 1 >> inds_ids.txt; done;
cut -f 2,3,4 ../geno3/update_ids.txt > tmp.txt;
paste inds_ids.txt tmp.txt > noid_update_ids.txt
tail noid_update_ids.txt
````

````bash
ls -1 *bim | sed -e 's/\.bim//g' > boxes.txt
if [ -e failed_sexcheck_ids ]; then rm failed_sexcheck_ids; fi;

while read box; do
     # remove family info
     n=`cat ${box}.fam | wc -l`;
     if [ -e inds_ids.txt ]; then rm inds_ids.txt; fi;
     for i in `seq 1 $n`; do echo 1 >> inds_ids.txt; done;
     cut -d' ' -f 2-6 ${box}.fam > fam_no_famids.txt;
     cp ${box}.fam ${box}.old_fam;
     paste -d ' ' inds_ids.txt fam_no_famids.txt > ${box}.fam

     plink --bfile ${box} --update-sex noid_update_sex.txt --make-bed --out $box;
     plink --bfile ${box} --update-ids noid_update_ids.txt --make-bed --out $box;
     plink --bfile ${box} --check-sex --out ${box};
     grep "PROBLEM" ${box}.sexcheck | awk '{print $1, $2}' >> failed_sexcheck_ids;
done < boxes.txt
````
